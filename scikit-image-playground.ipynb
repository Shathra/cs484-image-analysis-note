```python
import numpy as np
from skimage import data, io
import matplotlib.pyplot as plt

DPI = 96
```

```python
def show(im, size=None):
    if not size:
        size = (im.shape[1], im.shape[0])
    plt.figure(figsize=(size[0]/DPI, size[1]/DPI), dpi=DPI)
    io.imshow(im)
    io.show()

show(data.coins())
```

```python
im = data.coins()
im[50:52,:] = 0
show(im)

im = np.random.randint(0, 256, size=(720, 480), dtype='uint8')
show(im)
```

```python
# Images are simply numpy arrays
im = data.coins()
type(im)
im.shape
print(im)

y = np.histogram(im, bins=128)[0]
x = range(0, 256, 2)
x = range(128)
plt.figure(figsize=(8,5))
plt.bar(x, y, edgecolor = "none")
plt.show()

# Simple thresholding
im[im < 128] = 0
im[im >= 128] = 255
show(im)
```

```python
im = data.coffee()
show(im)
print(im.shape)

im = (im * 0.6).astype('uint8')

im[:, :, 0] = im[:, :, 0] // 2
show(im)
```
